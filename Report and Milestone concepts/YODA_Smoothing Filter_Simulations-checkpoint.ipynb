```julia
using PyPlot
input= [100,125, 148, 168, 184, 185, 200, 198, 190, 177, 159, 148, 113, 87, 63, 41, 23, 17, 2, 0, 5, 16, 32, 52, 75]

figure(figsize=(6.4,2.5));    # default is seems to be figsize=(6.4,5) 
plot(input,".-");
title("Noise Filter");

figure(figsize=(6.4,2.5));    # default is seems to be figsize=(6.4,5) 
output = [112,135,156,171,184,192,193,191,181,169,149,127,103,76,54,36,21,11,6,6,13,26,44,65,88]
plot(output,".-");
title("Smooth Filter");

# Both filters on same axis
figure(figsize=(6.4,2.5)); 
plot(input,".-");
plot(output,".-");
title("Smooth and Noisy Filters on same axis");
```
